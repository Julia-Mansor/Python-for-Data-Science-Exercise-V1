## (a) Merge both data frames using StudentID. (1 mark)

```python
import pandas as pd

# Load both CSV files
students_df = pd.read_csv('students.csv')
scores_df = pd.read_csv('scores.csv')

# Merge the dataframes
merged_df = pd.merge(students_df, scores_df, on='StudentID', how='inner')
```

```python
# Create the students DataFrame
students_data = {
    'StudentID': [101, 102, 103, 104, 105],
    'Name': ['Ali', 'Siti', 'John', 'Mei', 'Kumar'],
    'Department': ['DS', 'AI', 'DS', 'CS', 'AI'],
    'CGPA': [3.70, 3.45, 3.92, 3.60, 3.10]
}
students_df = pd.DataFrame(students_data)

# Create the scores DataFrame
scores_data = {
    'StudentID': [101, 102, 103, 104, 105],
    'Subject': ['Python', 'Python', 'Python', 'Python', 'Python'],
    'Marks': [85, 80, 92, 88, 75]
}

scores_df = pd.DataFrame(scores_data)


# Merge the dataframes
merged_df = pd.merge(students_df, scores_df, on='StudentID', how='inner')
print(merged_df)
```

## (b) Calculate the average Marks for all students. (1 mark)

```python
# Calculate the average Marks for all students
average_marks = scores_df['Marks'].mean()
print(f"Average Marks for all students: {average_marks}")
```

## (c) Use the groupby() function to calculate the average Marks for each Department. (1 mark)

```python
# First, merge the dataframes to get Department information with Marks
merged_df = pd.merge(students_df, scores_df, on='StudentID', how='inner')

# Use groupby() to calculate average Marks for each Department
avg_marks_by_dept = merged_df.groupby('Department')['Marks'].mean()
print("Average Marks by Department:")
print(avg_marks_by_dept)
```

## (d) Create a Pivot Table showing the average Marks by Department. (1 mark)

```python
# Create a Pivot Table showing average Marks by Department
pivot_table = merged_df.pivot_table(values='Marks', index='Department', aggfunc='mean')
print("Pivot Table - Average Marks by Department:")
print(pivot_table)
```

# Question 2 (4 Marks)

```python
def student_grade(mark):
    """
    Function to determine student grade category based on marks
    """
    if mark >= 80:
        return "Excellent"
    elif mark >= 60:  # This covers marks between 60 and 79
        return "Good"
    else:  # This covers marks < 60
        return "Need Improvement"

# Given list of marks
marks = [85, 72, 91, 55, 67, 49]

# Display the category for every student
print("Student Grade Categories:")
for i, mark in enumerate(marks, 1):
    grade = student_grade(mark)
    print(f"Student {i}: Mark = {mark}, Grade = {grade}")
```

# Question 4 (4 Marks)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
marks = [65,70,72,75,76,78,80,82,85,88,
         90,91,93,95,97,55,58,60,62,99]

# Create the box plot
plt.figure(figsize=(8, 6))
box_plot = plt.boxplot(marks, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
plt.title('Box Plot of Students\' Examination Marks', fontsize=14, fontweight='bold')
plt.ylabel('Marks', fontsize=12)
plt.xlabel('Students', fontsize=12)
plt.grid(True, alpha=0.3, axis='y')

# Display the plot
plt.tight_layout()
plt.show()

# Calculate statistics for analysis
marks_sorted = sorted(marks)
q1 = np.percentile(marks, 25)
median = np.percentile(marks, 50)
q3 = np.percentile(marks, 75)
iqr = q3 - q1

print(f"Statistical Analysis:")
print(f"Sorted marks: {marks_sorted}")
print(f"Q1 (25th percentile): {q1}")
print(f"Median (50th percentile): {median}")
print(f"Q3 (75th percentile): {q3}")
print(f"IQR (Q3 - Q1): {iqr}")

# Check for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = [x for x in marks if x < lower_bound or x > upper_bound]

print(f"\nOutlier Analysis:")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

Based on the box plot analysis:

**i. Median**: Median is **79.0**. This is the middle line inside the box, representing the 50th percentile of the data.

**ii. Interquartile Range (IQR)**: The IQR is **21.5** (Q3 - Q1 = 90.25 - 68.75). This represents the range containing the middle 50% of the data, shown as the height of the box.

**iii. Outliers**: **No outliers exist** in this dataset. All marks fall within the acceptable range defined by the outlier boundaries:
- Lower bound: 36.5 (Q1 - 1.5 × IQR)
- Upper bound: 122.5 (Q3 + 1.5 × IQR)

Since all marks (ranging from 55 to 99) fall between 36.5 and 122.5, there are no statistical outliers according to the standard 1.5 × IQR rule. What might appear as extreme values (like 55 or 99) are actually within the normal variation expected for this dataset.

Box plot shows a relatively well distributed dataset with the median slightly below the center of the box, indicating a slight left skew in the distribution.

## (a) Divide the marks into three bins using pd.cut():

```python
# Data
marks = [45,55,62,68,72,75,81,83,88,91,
         59,64,70,76,85,90,95,48,52,66]

# Create a pandas Series from the marks
marks_series = pd.Series(marks)

# Define bins and labels
bins = [0, 59, 79, 100]
labels = ['Low (0-59)', 'Medium (60-79)', 'High (80-100)']

# Use pd.cut() to categorize marks into bins
marks_binned = pd.cut(marks_series, bins=bins, labels=labels, include_lowest=True)

# Display results
print("Original marks:")
print(marks)
print("\nMarks categorized into bins:")
print(marks_binned.tolist())

# Show frequency count for each category
print("\nFrequency count:")
print(marks_binned.value_counts().sort_index())

# Create a DataFrame for better visualization
df = pd.DataFrame({
    'Marks': marks,
    'Category': marks_binned
})
print("\nDetailed breakdown:")
print(df)
```

## (b) Draw a Pie Chart showing the percentage of students in each category.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Get frequency count for each category
category_counts = marks_binned.value_counts().sort_index()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts.values, 
         labels=category_counts.index,
         autopct='%1.1f%%',
         startangle=90,
         colors=['lightcoral', 'lightskyblue', 'lightgreen'])

plt.title('Distribution of Students\' Final Marks by Category', 
          fontsize=14, fontweight='bold', pad=20)

plt.axis('equal')  # Equal aspect ratio ensures pie chart is circular
plt.tight_layout()
plt.show()

# Display the actual counts and percentages
print("Category breakdown:")
for category, count in category_counts.items():
    percentage = (count / len(marks)) * 100
    print(f"{category}: {count} students ({percentage:.1f}%)")
```
